```python
# ul=q3+1.5*(IQR)
# ll=q1-1.5*(IQR)
# IQR=q3-q1
```

```python
# caluclating upper limit asnd lower limit for outliers
def out(col):
    q1,q3=np.percentile(col,[25,76])
    iqr=q3-q1
    lw=q1-1.5*(iqr)
    uw=q3+1.5*(iqr)
    return lw,uw

out(df['GDP'])
```

```python
# treating the outliers
for i in df.select_dtypes(include= 'number').columns:
    lw,uw=out(df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])

    
```

```python
df1=df.groupby(['Country','Status'])['Life expectancy '].agg(['mean']).reset_index()
df1
```

```python
df2=pd.merge(df,df1,how='outer',on=['Country','Status'])
df2
```

```python
# Feature selection
X=df2.drop(columns=['Country','Status','Life expectancy '])
y=df2['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(X)
```

```python
x_train_scaled,x_test_scaled,y_train_scaled,y_test_scaled=train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
print(x_train_scaled.shape,y_train_scaled.shape)
print(x_test_scaled.shape,y_test_scaled.shape)
```

```python
from sklearn.metrics import r2_score
```

```python
# Model selection
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train_scaled,y_train_scaled)
y_pred_lr_test=lr.predict(x_test_scaled)
y_pred_lr_train=lr.predict(x_train_scaled)
```

```python
print('training accuracy',r2_score(y_train_scaled,y_pred_lr_train))
print('testing accuracy',r2_score(y_test_scaled,y_pred_lr_test))
```

```python
svr=SVR()
svr.fit(x_train_scaled,y_train_scaled)
y_pred_sv_test=svr.predict(x_test_scaled)
y_pred_sv_train=svr.predict(x_train_scaled)
```

```python
print('training accuracy',r2_score(y_train_scaled,y_pred_sv_train))
print('testing accuracy',r2_score(y_test_scaled,y_pred_sv_test))
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train_scaled,y_train_scaled)
y_pred_knn_test=knn.predict(x_test_scaled)
y_pred_knn_train=knn.predict(x_train_scaled)
```

```python
print('training accuracy',r2_score(y_train_scaled,y_pred_knn_train))
print('testing accuracy',r2_score(y_test_scaled,y_pred_knn_test))
```

```python
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred_dt_test=dt.predict(x_test)
y_pred_dt_train=dt.predict(x_train)
```

```python
print('training accuracy',r2_score(y_train,y_pred_dt_train))
print('testing accuracy',r2_score(y_test,y_pred_dt_test))
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf_test=rf.predict(x_test)
y_pred_rf_train=rf.predict(x_train)
```

```python
print('training accuracy',r2_score(y_train,y_pred_rf_train))
print('testing accuracy',r2_score(y_test,y_pred_rf_test))
```

```python
et=ExtraTreesRegressor()
et.fit(x_train,y_train)
y_pred_et_test=et.predict(x_test)
y_pred_et_train=et.predict(x_train)
```

```python
print('training accuracy',r2_score(y_train,y_pred_et_train))
print('testing accuracy',r2_score(y_test,y_pred_et_test))
```
