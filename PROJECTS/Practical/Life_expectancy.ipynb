```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Life Expectancy Data.csv')
df
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
df.columns
```

```python
df.duplicated().sum()
```

```python
for i in [' BMI ','Polio','Income composition of resources']:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include='number').columns:
    df[i]=impute.fit_transform(df[[i]])
```

```python
df.isnull().sum().sum()
```

```python
col=df.columns
col
```

```python
def whisker(col):
    global lw,uw
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*(iqr)
    uw=q3+1.5*(iqr)
    return lw,uw
whisker(df['GDP'])
```

```python
for i in df.select_dtypes(include='number').columns:
    lw,uw=whisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
for i in ['GDP','Total expenditure',' thinness  1-19 years',' thinness 5-9 years']:
    plt.boxplot(df[i])
    plt.show()
```

```python
(df['Country'].value_counts().values==16).sum()
```

```python
df1=df.groupby(['Country','Status'])['Life expectancy '].mean().reset_index()
df1
```

```python
df2=pd.merge(df,df1,how='inner',on=['Country','Status'])
df2
```

```python
df2=pd.get_dummies(data=df2,columns=['Status'],drop_first=True)
df2
```

```python
for i in df2.select_dtypes(include='number').columns:
    plt.boxplot(df2[i])
    plt.show()
```

```python
df2.columns
```

```python
X=df2.drop(columns=['Country','Life expectancy _x'])
y=df2['Life expectancy _x']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error as mse
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=2)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_train=lr.predict(x_train)
y_pred_test=lr.predict(x_test)
```

```python
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```

```python
mse(y_train,y_pred_train)
```

```python
mse(y_test,y_pred_test)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
y_pred_train=svr.predict(x_train)
y_pred_test=svr.predict(x_test)
```

```python
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```
