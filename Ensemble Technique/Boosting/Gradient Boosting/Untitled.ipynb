```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
df=load_iris()
df
```

```python
X=df.data
```

```python
y=df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
ada_clf=AdaBoostClassifier(n_estimators=50,learning_rate=0.5,random_state=42)
```

```python
ada_clf.fit(x_train,y_train)
```

```python
y_pred_train1=ada_clf.predict(x_train)
y_pred_test1=ada_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('training accuracy',accuracy_score(y_train,y_pred_train1))
print('testing accuracy',accuracy_score(y_test,y_pred_test1))
```

# Hyperparameter tuning on Adaboost

```python
param_grid={'n_estimators':[50,100,150],
'learning_rate':[0.01,0.1,0.5,1],}
 


    
```

```python
grid_search=GridSearchCV(estimator=ada_clf,param_grid=param_grid,cv=5,scoring='accuracy',n_jobs=-1)
```

```python
import time
```

```python
start=time.time()
grid_search.fit(x_train,y_train)
end=time.time()
print(end-start)
```

```python
best_paramters=grid_search.best_params_
best_paramters
```

```python
ada_clf1=AdaBoostClassifier(**best_paramters)
```

```python
ada_clf1.fit(x_train,y_train)
```

```python
Y_pred_train1=ada_clf1.predict(x_train)
Y_pred_test1=ada_clf1.predict(x_test)
```

```python
print('training accuracy',accuracy_score(y_train,Y_pred_train1))
print('testing accuracy',accuracy_score(y_test,Y_pred_test1))
```
