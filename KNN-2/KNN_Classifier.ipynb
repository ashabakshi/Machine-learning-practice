```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df
```

```python
df['diagnosis'].replace({'M':1,'B':0},inplace=True)
df
```

```python
X=df.drop(columns=['id','diagnosis','Unnamed: 32'])
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(X)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
print(x_train.shape,y_train.shape)
```

```python
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# !pip install --upgrade threadpoolctl scikit-learn
```

```python
y_pred_train=knn.predict(x_train)
```

```python
y_pred_test=knn.predict(x_test)
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
scores={}
scores_t={}

for i in range(1,16):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    y_pred_train=knn.predict(x_train)
    scores[i]=accuracy_score(y_test,y_pred)
    scores_t[i]=accuracy_score(y_train,y_pred_train)

print(scores)
print(scores_t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(scores.keys(),scores.values(),marker='o',label='test')
plt.plot(scores_t.keys(),scores_t.values(),marker='o',color='k',label='train')
plt.legend();
```
