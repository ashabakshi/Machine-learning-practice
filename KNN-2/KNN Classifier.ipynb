```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df
```

```python
df.isna().sum()
```

```python
df['diagnosis'].replace({'M':1,'B':0},inplace=True)
```

```python
df.head(2)
```

```python
X=df.drop(columns=['id','diagnosis','Unnamed: 32'])
y=df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
# x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.predict(x_train)
```

```python
x_test
```

```python
y_pred_test=knn.predict(x_test)
```

```python
y_pred_train=knn.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

# How to select K

```python
score_train={}
test=[]
for i in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred_test=knn.predict(x_test)
    y_pred_train=knn.predict(x_train)
    score_train[i]=accuracy_score(y_train,y_pred_train)
    test.append(accuracy_score(y_test,y_pred_test))



print(score_train)
    
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(score_train.keys(),score_train.values(),marker='o',color='yellow',label='train')
plt.plot(score_train.keys(),test,marker='o',label='test')
plt.xticks([i for i in range(1,21)])
plt.legend();
```
