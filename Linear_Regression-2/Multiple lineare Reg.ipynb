```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('car data.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Age']=2025-df['Year']
df
```

```python
df.drop(columns=['Year'],inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission'],drop_first=True)
df
```

```python
df
```

```python
# df['Car_Name'].nunique()
```

```python
df1 = df.groupby(['Car_Name'])['Selling_Price'].mean().reset_index().rename(columns = {'Selling_Price':'avg_price'})
df1
```

```python
df1 = pd.merge(df,df1,on='Car_Name',how='inner')
df1
```

```python
df1.drop(columns=['Car_Name'],inplace=True)
```

```python
df1
```

```python
sns.heatmap(df.corr(),annot=True);
```

```python
X = df1.drop(columns=['Selling_Price'])
y=df1['Selling_Price']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
x_scaled = st.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape,y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred_train = lr.predict(x_train)
```

```python
y_pred_test = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
print(mean_squared_error(y_train,y_pred_train))
print(mean_squared_error(y_test,y_pred_test))
```

```python
r2_score(y_train,y_pred_train)
```

```python
r2_score(y_test,y_pred_test)
```
